```python
def get_name_with_age(name: str, age: int) -> str:
    name_age = "The name is " + name + " has age " + str(age) 
    return name_age

get_name_with_age("Tj", 30)
```

```python
from datetime import datetime
from pydantic import BaseModel

class User(BaseModel):
    id: int
    name: str = "tj"
    age: int | None
    signup_ts: datetime | None = None
    friends: list[int] = []

external_data = {
    "id": "123",
    "signup_ts": "2017-06-01 12:22",
    "friends": [1, "2", b"3"],
    "age": "54"
}

user = User(**external_data)
print(user)
print(user.id)
```

```python
from typing import Annotated

def say_hi(name: Annotated[str, "The is the metadata"]) -> str:
    return f"hello {name}"

say_hi("tj")
```
