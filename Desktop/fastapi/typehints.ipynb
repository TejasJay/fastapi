```python
from datetime import datetime
from pydantic import BaseModel

class User(BaseModel):
    id: int
    name: str = "tj"
    age: int | None
    signup_ts: datetime | None = None
    friends: list[int] = []

external_data = {
    "id": "123",
    "signup_ts": "2017-06-01 12:22",
    "friends": [1, "2", b"3"],
    "age": "54"
}

user = User(**external_data)
print(user)
print(user.id)
```

```python
from typing import Literal, Type, Dict
from datetime import date
from pydantic import BaseModel, Field, create_model
from fastapi import HTTPException
```

```python
CATEGORY_DEFINITIONS = {
    1: {"name": "Laptop",
        "fields": {"cpu_type": (str, ...), "ram_gb": (int, ...)}},
    2: {"name": "T-Shirt",
        "fields": {"color": (str, ...), "size": (Literal['S','M','L','XL'], ...)}},
    3: {"name": "Equipment",
        "fields": {"voltage": (int, 220), "warranty_expires_on": (date, ...)}}
}
```

```python
def get_product_model_for_category(categoty_id: int) -> BaseModel:
    category = CATEGORY_DEFINITIONS[categoty_id]
    if not category:
        raise HTTPException(status_code=404, detail=f"Product category {categoty_id} not found")
    
    base_fields = {
        "sku": (str, ...),
        "price": (float, Field(..., ge=0))
    }

    all_fields = {**base_fields, **category["fields"]}

    ProductModel = create_model(
        f"Dynamic{category['name']}Model",
        **all_fields
    )
    return ProductModel
```

```python
get_product_model_for_category(3)
```
